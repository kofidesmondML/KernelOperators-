```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def rbf(x, y, epsilon=1.0):
    return np.exp(- (epsilon**2) * (x**2 + y**2))
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
eps=[0.5,1.0,3.0, 5.0,10,100]
for epsilon in eps:
    Z = rbf(X, Y, epsilon)
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('RBF Value')
    ax.set_title(f'3D Plot of a Radial Basis Function (ε = {epsilon})')
    plt.show()
```
